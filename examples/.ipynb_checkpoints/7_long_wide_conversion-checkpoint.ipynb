```python
import pandas as pd
import numpy as np
```

Demonstrate how to convert from wide to long, then go back to wide form

```python
# prepare a wide form dataframe
np.random.seed(42)
data1 = np.random.randn(10)
data2 = np.random.randn(10)
data3 = np.random.randn(10)
wide = pd.DataFrame({'var1':data1,'var2':data2,'var3':data3})   # wideform
```

```python
wide
```

```python
# very important, add an identity column, otherwise, we lose the track of each value. In anothe word, in this wide 
# form data frame, index 0 and column "var1" will determine a unique and unambiguous entry. We want to keep that.
wide = wide.reset_index(level=-1)
```

```python
wide
```

```python
# convert to long, by using melt function
long = pd.melt(wide,id_vars='index',value_vars=['var1','var2','var3'],var_name='var',value_name='values')
```

```python
long
```

```python
# go back to wide, using pivot function
back = pd.pivot(long,index=['index'],columns=['var'],values=['values'])
```

```python
back
```

```python
# the column become a MultiIndex, we don't need it, reset it to a normal Index
back.columns = pd.Index(['var1','var2','var3'])
```

```python
back
```
